```python
# -- IMPORTS -- #
import math
import scipy
import numpy as np
from skimage import util
import matplotlib.pyplot as plt
from keras.datasets import mnist
from keras.utils import to_categorical
```

```python
# -- TRAIN AND TEST DATA PREPARATION -- #
(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

```python
# -- CONVERTING DATA FOR INPUT INTO NEURAL NETWORK -- #
x_train = x_train.reshape((60000,28*28))
x_train = x_train.astype("float32") / 255

x_test = x_test.reshape((10000,28*28))
x_test = x_test.astype("float32") / 255

y_train = to_categorical(y_train)
y_test = to_categorical(y_test)
```

```python
# -- DIMENSIONALITY REDUCTION -- #
import dim_reduction
x_train0, x_test0 = x_train.copy(), x_test.copy()
#x_train1, x_test1 = dim_reduction.apply(100, x_train, x_test, "FA")
x_train2, x_test2 = dim_reduction.apply(50, x_train, x_test, "FA")
x_train3, x_test3 = dim_reduction.apply(40, x_train, x_test, "FA")
x_train4, x_test4 = dim_reduction.apply(30, x_train, x_test, "FA")
x_train5, x_test5 = dim_reduction.apply(20, x_train, x_test, "FA")
x_train6, x_test6 = dim_reduction.apply(10, x_train, x_test, "FA")
```

```python
# -- NEURAL NETWORK -- #
import neural_network
history0 = neural_network.fit(784, 30, x_train0, y_train, x_test0, y_test)
#history1 = neural_network.fit(100, 30, x_train1, y_train, x_test1, y_test)
history2 = neural_network.fit(50, 30, x_train2, y_train, x_test2, y_test)
history3 = neural_network.fit(40, 30, x_train3, y_train, x_test3, y_test)
history4 = neural_network.fit(30, 30, x_train4, y_train, x_test4, y_test)
history5 = neural_network.fit(20, 30, x_train5, y_train, x_test5, y_test)
history6 = neural_network.fit(10, 30, x_train6, y_train, x_test6, y_test)
```

```python
# -- OUTPUT PLOTS -- #
fig, ax = plt.subplots()
plt.minorticks_on()
plt.plot(history0.history["val_accuracy"])
#plt.plot(history1.history["val_accuracy"])
plt.plot(history2.history["val_accuracy"])
plt.plot(history3.history["val_accuracy"])
plt.plot(history4.history["val_accuracy"])
plt.plot(history5.history["val_accuracy"])
plt.plot(history6.history["val_accuracy"])

plt.title("Neural Network Performance on Clean MNIST with FA")
plt.ylabel("Validation Accuracy")
plt.ylim(0.8, 1.0)

ax.yaxis.grid(which="major", color='#666666', linestyle='-', alpha=0.8)
ax.yaxis.grid(which="minor", color='#999999', linestyle='-', alpha=0.2)

plt.xlabel("Number of Epochs")
plt.legend(["n=784", "n=50",  "n=40", "n=30", "n=20", "n=10"],loc="upper left",ncol=3,fontsize = 'small')
plt.savefig('nn_clean_fa_var-ncomp.png')
plt.show()
```
